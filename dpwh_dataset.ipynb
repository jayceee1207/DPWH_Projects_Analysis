```python
#transform parquet file to csv

import pandas as pd
parquet_file_path = 'dpwh_transparency_data.parquet'
csv_output_path = 'dpwh_transparency_data.csv'
df = pd.read_parquet(parquet_file_path)
df.to_csv(csv_output_path, index=False)
print(f"Successfully converted '{parquet_file_path}' to '{csv_output_path}'")
```
